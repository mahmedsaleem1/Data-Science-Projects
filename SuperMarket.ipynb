# Data Visualization and Analysis

#### How do sales vary across different branches and cities?

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Branch', y='Total', hue='City', data=data)
plt.title('Sales Variation Across Branches and Cities',color='red')
plt.xlabel('Branch',color='blue')
plt.ylabel('Total Sales',color='blue')
plt.legend(title='City')
plt.show()
```

##### Observation : Order of sales(Branch 'C' --> Branch 'B' --> Branch 'A')

#### What is the distribution of customer types across different branches or cities?

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Branch', hue='Customer type', data=data)
plt.title('Distribution of Customer Types Across Branches',color='red')
plt.xlabel('Branch',color='blue')
plt.ylabel('Count',color='blue')
plt.legend(title='Customer Type')
plt.show()
```

##### Observation :  
#####    Branch 'A' --> More Normal Customers than Members.
#####    Branch 'B' --> More Members than Normal Customers.
#####    Branch 'C' --> Members and Normal Customer were almost equal but a bit more Normal Customers.

#### How does the unit price of products vary across different product lines?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product line', y='Unit price', data=data)
plt.title('Unit Price Variation Across Product Lines',color='red')
plt.xlabel('Product Line',color='blue')
plt.ylabel('Unit Price',color='blue')
plt.xticks(rotation=45)
plt.show()
```

#### Are there any peak times for sales during the day or specific days of the week?

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Hour', y='Total', data=data)
plt.title('Hourly Sales Variation',color='red')
plt.xlabel('Hour of the Day',color='blue')
plt.ylabel('Total Sales',color='blue')
plt.show()
```

#### Are there any peak times for sales during the day or specific days of the week?

```python
data['Month'] = data['Month'].astype(int)

monthly_sales = data.groupby('Month')['Total'].sum()

plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title('Monthly Sales Variation',color='red')
plt.xlabel('Month',color='blue')
plt.ylabel('Total Sales',color='blue')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

##### Observation : Sales only in the first three months, most sales during 1st month and least sale during 2nd month
